```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import numpy as np

T_MAX = 1000
W_SIZE = 1000

def init_random():
    world = np.zeros((T_MAX,W_SIZE)).astype(np.int)
    world[0] = ((np.random.random(W_SIZE) > 0.5 ) *1.0)
    world[0,0:100] = 0
    world[0,900:] = 0

    return world

def init_one():
    world = np.zeros((T_MAX,W_SIZE)).astype(np.int)
    world[0][W_SIZE//2] = 1
    return world

# 1 0 1 -> 1
# 0 1 1 -> 0

# 30, 110, 102 (xor)
def set_rule(r=102):
    rule = {
        ((i>>2)&1, (i>>1)&1, (i>>0)&1): (r>>i)&1 for i in range(8)}
    return rule
```

```python
def do():
    N = len(world)
    for time in range(T_MAX-1):
        for cell in range(W_SIZE):
            if cell == 0:
                continue
            if cell == W_SIZE - 1:
                continue
            neighbors = tuple(world[time][cell-1:cell+2])
            # print(time, cell, "TS")
            world[time+1][cell] = rule[neighbors]
# do()
```

```python
world = init_one() #init_random()
rule = set_rule()
do()
plt.imshow(world)
```

```python
world = init_random()
rule = set_rule()
do()
plt.imshow(world)
```
